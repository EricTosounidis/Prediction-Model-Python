```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC") # Ticker class -> enable us to download price history
```

```python
sp500 = sp500.history(period="max") # query historical prices, max = from beginning to end data
```

```python
sp500 # we get a data frame where each row is the price of each trading day
```

```python
sp500.index # looking at the date time index(leftest column)
```

```python
# Cleaning Data and Visualizing it
```

```python
sp500.plot.line(y="Close", use_index=True)  # plotting the closing price against the index
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
# Setting up our target
```

```python
sp500.loc[:, "Tomorrow"] = sp500["Close"].shift(-1) # showing tomorrow's price
```

```python
sp500
```

```python
sp500.loc[:, "Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int) # Setting up a target(is tomorrow's price > than today's price)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy() # using loc method from pandas to take a better date to data set
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1) # number of decision trees we want to train for a better accuracy, (no linear correlation between Open and Tomorrow's price)

train = sp500.iloc[:-100] # Creating a baseline model to split data 
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"]) # Using predictor columns to predict the target
```

```python
# Measuring how precise our model is
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd
preds = pd.Series(preds, index = test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
# Build a backtesting system
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts() # How many days we predicted the market goes up or down
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# Percentage of days the market went up to compare
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# Adding more predicors to our model to improve our accuracy
```

```python
horizons = [2,5,60,250,1000] # calculate the mean of close prices the last 2/5/60/250/1000 days

new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500.loc[:, ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500.loc[:, trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna(subset=sp500.columns[sp500.columns != "Tomorrow"])
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=500, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1] # proba = will return a probability of the stock going down or up
    preds[preds >= .6] = 1 # creating a threshold for the model to be more confident if the price will go up
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```
